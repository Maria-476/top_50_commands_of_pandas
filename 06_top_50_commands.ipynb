# Step-2 Handling Missing Values

### Drop rows with any missing values

```python
df.dropna()
df.head()
df['deck'].dropna
df.head()
df.drop(columns='deck',inplace=True)
df.head()
```

### Drop specific rows with missing values

```python
df.dropna(subset=['age' , 'embarked']) # just to show
# it will rwmovw whole rows with missing values in the columns of age and embarked
```

### Fill missing values with a fixed value

```python
df.fillna(0)
# it does not work because embarked and embark _town are not numeric columns
```

### Fill missing value with median

```python
df['age'].median()
```

```python
df['age'] = df['age'].fillna(df['age'].median())

# to update DataFrame is better instead of creating a new one or inplace=True method
```

```python
df.isnull().sum() # to check the number of missing values in 'age' column
```

### Fill missing values with mode

```python
df['embark_town'].unique()
```

```python
df['embark_town'].value_counts()
```

```python
df['embarked'] = df['embarked'].fillna(df['embarked'].mode()[0])
df['embark_town'] = df['embark_town'].fillna(df['embark_town'].mode()[0])
df
```

```python
df.isnull().sum() # just to check
```

### Interpolate missing values

```python
df.interpolate()
# Interpolation is the process of estimating an unknown value that lies between two known values.
# interpolation is used only for numerical values not for categorical values
# df['age'].interpolate()

```
